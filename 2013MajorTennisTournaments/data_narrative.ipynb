```python
break_points_created=np.zeros(len(aus_open_women_players))
break_points_won=np.zeros(len(aus_open_women_players))
conversion_rate=np.zeros(len(aus_open_women_players))
j=0
for player in aus_open_women_players['Name of Player']:
    break_point_created=0
    break_point_won=0
    aus_open_women_player_wise=ausopen_women.where(np.logical_or(ausopen_women['Player1']==player,ausopen_women['Player2']==player)).dropna(how='all')
    for i in range(len(aus_open_women_player_wise)):
        if np.array(aus_open_women_player_wise['Player1'])[i]==player:
            break_point_created+=aus_open_women_player_wise['BPC.1'].iloc[i]
            break_point_won+=aus_open_women_player_wise['BPW.1'].iloc[i]
        else:
            break_point_created+=aus_open_women_player_wise['BPC.2'].iloc[i]
            break_point_won+=aus_open_women_player_wise['BPW.2'].iloc[i]
        if break_point_created==0:
            conv_rate=0
        else:
            conv_rate=break_point_won/break_point_created
    break_points_created[j]=break_point_created
    break_points_won[j]=break_point_won
    conversion_rate[j]=conv_rate
    j+=1
aus_open_women_players['Break points created']=break_points_created
aus_open_women_players['Break points won']=break_points_won
aus_open_women_players['Conversion rate']=conversion_rate
aus_open_women_players
```
